### DL condition

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys, os


current_dir = os.getcwd()  
project_root = os.path.abspath(os.path.join(current_dir, '..', '..'))  # goes two levels up to MOSAIC
sys.path.append(project_root)

from configs.dreamachine2 import config

sanitized_model_name = config.transformer_model.replace('/', '_') #name of the transformer model with '/' replaced by '_'
file_path = f'OPTUNA_results_DL_sentences_{sanitized_model_name}.csv'
df = pd.read_csv(file_path)


df['best_embedding_coherence'] = df['embedding_coherence'].cummax()

plt.figure(figsize=(12, 7))
plt.plot(df['trial_number'], df['best_embedding_coherence'], marker='.', linestyle='-', color='b', alpha=0.7)

plt.title('Optuna Optimization History (DL)', fontsize=16)
plt.xlabel('Trial Number', fontsize=12)
plt.ylabel('Best Embedding Coherence Found So Far', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# plt.savefig('optimization_history_DL.png')

plt.show()
```
